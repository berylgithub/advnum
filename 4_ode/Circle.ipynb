```python
'''
circle with radius = 1
'''

import numpy as np
from matplotlib import pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
radius = 1
x = radius*np.cos(theta)
y = radius*np.sin(theta)
length = len(x)

# test on plot
plt.plot(x, y)
plt.axis('equal')
plt.title('Circle')
plt.show()

# generate segment idx:
segment_start = np.arange(1, length+1, 1)
segment_end = np.arange(2, length+2, 1)
segment_end[-1] = 1
print(segment_start, segment_end, len(segment_start)==len(segment_end))

# print (x,y) to txt according to triangle's format
with open('circle.poly', 'w') as f:
    f.writelines("#circle with radius = 1, 100 points \n")
    f.writelines(str(len(x))+" 2 0 0\n")
    for i, _ in enumerate(x):
        f.write(str(i+1)+" "+str(x[i])+" "+str(y[i])+" 0\n")
    f.write("# circle segments:\n")
    f.write(str(length)+" 0\n")
    for i in range(length):
        f.write(str(i+1)+" "+str(segment_start[i])+" "+str(segment_end[i])+" 0\n")
    f.write("#hole:\n")
    f.write("0\n")
```

```python
'''
circle(r=2) with another circle(r=1) within
'''
radius = 2
x_out = radius*np.cos(theta)
y_out = radius*np.sin(theta)

# test on plot
plt.plot(x, y, label="inner")
plt.plot(x_out, y_out, label="outer")
plt.axis('equal')
plt.title('Circle')
plt.show()



# print (x,y) to txt according to triangle's format
with open('donut.poly', 'w') as f:
    f.writelines("#donut with outer radius = 2, inner radius = 1, 100 points each \n")
    f.writelines(str(length*2)+" 2 0 0\n")
    i=0
    f.write("# inner circle:\n")
    while i<length: # iner circle:
        f.write(str(i+1)+" "+str(x[i])+" "+str(y[i])+" 0\n")
        i+=1
    i=0
    f.write("# outer circle:\n")
    while i<length:
        f.write(str(i+1+length)+" "+str(x_out[i])+" "+str(y_out[i])+" 0\n")
        i+=1
    i=0
    f.write("# inner circle segments:\n")
    f.write(str(length*2)+" 0\n")
    while i<length:
        f.write(str(i+1)+" "+str(segment_start[i])+" "+str(segment_end[i])+" 0\n")
        i += 1
    i = 0
    f.write("# outer circle segments:\n")
    while i<length:
        f.write(str(i+1+length)+" "+str(segment_start[i]+length)+" "+str(segment_end[i]+length)+" 0\n")
        i += 1
    i = 0
    f.write("# hole :\n") 
    f.write("1\n")
    f.write("1 0 0") #hole in the centre
```
